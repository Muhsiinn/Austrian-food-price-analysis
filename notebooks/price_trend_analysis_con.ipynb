```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

sns.set(style="whitegrid")
```

```python
consumer_df = pd.read_csv("../data/consumer-price-indices_aut.csv")
consumer_df = consumer_df.iloc[1:]
consumer_df = consumer_df[["Year","Item","Value"]]
consumer_df.head()
```

```python
consumer_df["Item"].unique()
```

```python
consumer_df= consumer_df[consumer_df["Item"]=="Consumer Prices, Food Indices (2015 = 100)"]
consumer_df.head(-5)
```

```python
consumer_df["Year"] = consumer_df["Year"].astype(int)
consumer_df["Value"] = consumer_df["Value"].astype(float)
```

```python
#consumer_df["Normalized"] = consumer_df["Value"].apply(lambda x : (x - consumer_df["Value"].min())/(consumer_df["Value"].max() - consumer_df["Value"].min())  )
```

```python
consumer_df.head(30)
```

```python
# Step 1: Convert Year to int (if not already)
consumer_df["Year"] = consumer_df["Year"].astype(int)

# Step 2: Group by year and take average of monthly values
yearly_consumer = (
    consumer_df.groupby("Year")["Value"]
    .mean()
    .reset_index()
)

# Step 3: Normalize the yearly values
yearly_consumer["Normalized"] = (
    (yearly_consumer["Value"] - yearly_consumer["Value"].min())
    / (yearly_consumer["Value"].max() - yearly_consumer["Value"].min())
)
```

```python
yearly_consumer.head(100)
```

```python
print("Max value:", yearly_consumer["Value"].max())
print("Row with max value:")
print(yearly_consumer[yearly_consumer["Value"] == yearly_consumer["Value"].max()])

print("Max normalized value:", yearly_consumer["Normalized"].max())
```

```python
yearly_consumer.to_csv("normalized_consumer_prices.csv", index=False)

```
